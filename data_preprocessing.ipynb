```python
import pandas as pd
import re
import numpy as np



# Define the function to convert price strings to integers
def convert_price_to_int(price_str):
    # Remove the '₹', commas, and any text after the price
    cleaned_price = re.sub(r'[₹,]', '', price_str)  # Remove ₹ and commas
    cleaned_price = re.sub(r'[^0-9]', '', cleaned_price)  # Keep only numbers
    if cleaned_price:  # Check if the cleaned string is not empty
        return int(cleaned_price)
    else:
        return np.nan  # Return NaN if there's no valid price

# Apply the function to the price column
df['price for two'] = df['price for two'].apply(convert_price_to_int)

# Save the modified DataFrame to a CSV file
df.to_csv('cleaned_dataset.csv', index=False)


```

```python


# Function to extract opening and closing times safely
def extract_timings(timing):
    try:
        parts = timing.split(',')
        open_times = []
        close_times = []
        
        for part in parts:
            if '–' in part:  # Ensure there's a valid separator
                times = part.split('–')
                open_times.append(times[0].strip())
                close_times.append(times[1].split('(')[0].strip())
        
        # If no valid time was extracted, return None
        if not open_times or not close_times:
            return None, None
        
        # Return the first opening time and last closing time
        return open_times[0], close_times[-1]
    
    except IndexError:
        return None, None

# Apply the function to the 'opening & closing time' column
df['Opening Times'], df['Closing Times'] = zip(*df['opening & closing time'].apply(extract_timings))

df.head()
```

```python
df=df.drop(columns=["opening & closing time"])
df.to_csv('cleaned_dataset.csv', index=False)
```

```python
from geopy.geocoders import Nominatim
import time

# Initialize geolocator
geolocator = Nominatim(user_agent="restaurant_locator")

# Function to get coordinates
def get_coordinates(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        return None, None

# Assuming 'address' is a column in your DataFrame
df['latitude'] = None
df['longitude'] = None

for idx, row in df.iterrows():
    address = row['address']  # Replace with the actual column name for the address
    lat, lon = get_coordinates(address)
    df.at[idx, 'latitude'] = lat
    df.at[idx, 'longitude'] = lon
    
    time.sleep(1)  # Sleep for a second to avoid overwhelming the API


```

```python
import folium

# Initialize the map at a central location (e.g., Hyderabad)
map = folium.Map(location=[17.385044, 78.486671], zoom_start=12)

# Add markers for each restaurant
for _, row in df.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        folium.Marker([row['latitude'], row['longitude']], popup=row['names']).add_to(map)

# Save the map as an HTML file
map.save('restaurants_map.html')
```
